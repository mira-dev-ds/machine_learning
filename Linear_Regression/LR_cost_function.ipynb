## To implement cost function for Linear regression model with one variable


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_cost(x,y,w,b):
  m = x.shape[0]
  cost_sum = 0
  for i in range(m):
    f_wb = w * x[i] + b
    cost = (f_wb - y[i]) ** 2
    cost_sum += cost
  total_cost = (1/(2*m)) * cost_sum
  return total_cost
```

### Visualizing the Cost Function (Contour Plot)
To better understand the cost function, we can create a contour plot. This plot will show how the cost changes as we vary the parameters `w` and `b`.

W → grid of w values

B → grid of b values

Z → cost value for each (w, b) pair

Z[i][j] = J(W[i][j], B[i][j])


```python
# Define a range of w and b values for the contour plot
w_values = np.linspace(-100, 400, 100) # 100 values between -100 and 400
b_values = np.linspace(-200, 400, 100) # 100 values between -200 and 400

# Create a meshgrid from w and b values
W, B = np.meshgrid(w_values, b_values)

# Calculate the cost (Z) for each (w, b) pair
Z = np.zeros_like(W)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        Z[i, j] = compute_cost(x_train, y_train, W[i, j], B[i, j])
```

```python
# Create a 2D contour plot
plt.figure(figsize=(8, 6))

contour = plt.contourf(W, B, Z, levels=50, cmap='viridis')

#filled contour plot , Shows same-cost regions using colors
#divides cost into 50 color bands
#color scheme (dark = low cost, bright = high cost)

plt.colorbar(contour, label='Cost')
plt.xlabel('w')
plt.ylabel('b')
plt.title('Contour Plot of Cost Function')

# Highlight the optimal values (from previous output, w=200, b=100 leads to cost=0)
plt.scatter(200, 100, marker='x', color='red', s=100, label='Optimal (w=200, b=100)')
plt.legend()
plt.show()
```
