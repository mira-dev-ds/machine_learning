```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[2,2],[4,4],[6,6],[8,8],[1,7],[2,6]])
```

```python
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.title("Original Data")
plt.show()
```

```python
k = 2
cluster_centre = X[np.random.choice(len(X),k,replace = False)]
```

```python
for _ in range(100):

  distances = []

  for centre in cluster_centre:
    d = np.linalg.norm(X - centre, axis = 1)
    distances.append(d)

  distances = np.array(distances)
  labels = np.argmin(distances, axis = 0)

  new_cluster_centres = []

  for i in range(k):
    points = X[labels == i]
    new_cluster_centres.append(np.mean(points, axis = 0))

  new_cluster_centres = np.array(new_cluster_centres)

  if np.all(cluster_centre == new_cluster_centres):
    break
  cluster_centre = new_cluster_centres
```
